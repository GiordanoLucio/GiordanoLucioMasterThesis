```python
from shutil import copyfile
#copyfile(src, dst)
import xml.etree.ElementTree as ET
import os
```

```python
#remove all the annotations that are not desired, just add clause to remove more or less objects
for element in os.listdir("annotations"):
    print(element)
    fileurl="annotations/"+element
    fileurl2="1_Ann_person/"+element
    tree = ET.parse(fileurl)
    root = tree.getroot()
    num_person = 0
    for c in root.findall('object'):
        rank = c.find('name').text
        a = 'person' not in rank
        #b = 'chair' not in rank
        #if you want to keep more labels, just add it to the and clause
        #if (a and b):
        if(a):
            print('removing: ',rank)
            root.remove(c)
        elif 'person' in rank:
            print("found a person: ", rank)
            num_person += 1
        if num_person > 0:
            path=fileurl2
            tree.write(fileurl2)
            imgName = element[:-4]+'.jpg'
            print("Img name: ",imgName)
            copyfile('JPEGImages/'+imgName,"1_JPEGImages_person/"+imgName)
    #copyfile(src, dst)
    
```
